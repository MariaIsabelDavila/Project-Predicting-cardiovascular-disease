## Changing the null values

```python
#We replace the null values using the mean technique
cardio['Fruit Consumption'].fillna(cardio['Fruit Consumption'].mean(), inplace=True)
print(cardio['Fruit Consumption'].unique())
```

```python
# With simple random hot deck method, we replace the null values
import numpy as np
for i, row in cardio.iterrows():
    if row.isnull().any():
        neighbors=cardio.iloc[i-1:i+2].dropna()
        for col in cardio.columns:
            if pd.isnull(row[col]):
                cardio.at[i,col]=neighbors.sample(1)[col].values[0]
cardio
```

```python
from scipy import stats
ll_model_A = modelo_A.llf  # Log-likelihood of model A
ll_model_B = modelo_B.llf  # Log-likelihood of model A

# Calculate the difference in log-likelihoods
likelihood_ratio = 2 * (ll_model_B - ll_model_A)

# Degrees of freedom: Difference in the number of parameters between the models
df = abs(modelo_B.df_resid - modelo_A.df_resid)  

# Realizar el test de razón de verosimilitudes
p_value = 1 - stats.chi2.cdf(likelihood_ratio, df)

# Perform the likelihood ratio test
print(f"Likelihood Ratio: {likelihood_ratio}")
print(f"Degrees of Freedom: {df}")
print(f"P-Value: {p_value}")
```
